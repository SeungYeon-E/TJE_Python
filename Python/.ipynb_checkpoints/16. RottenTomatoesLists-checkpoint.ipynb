```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd

url = "https://www.rottentomatoes.com/top/bestofrt/?year=2019"
res = urlopen(url)
soup = BeautifulSoup(res, "html.parser")

# li_lists = soup.select("#top_movies_main > div > table > tr > td > a") 
li_lists = soup.select("#top_movies_main > div > table > tr > td > a") 

count = 0
titleLists = [] #영화제목
countLists = [] #예매율 Lists

for li in li_lists:
    count += 1
    print(count, li.string.strip())
    countLists.append(count)
    titleLists.append(li.string.strip())
```

```python
# List 확인
print(countLists)
print(titleLists)
```

```python
# countLists 만으로 DataFrame 생성
df = pd.DataFrame(data=countLists, columns=['Index'])
df['Titles'] = pd.Series(titleLists)
df
```

```python
# cvs로 저장
df.to_csv("./Data/MoviesList.csv", sep=",", encoding='utf-8', index=False)
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd

count = 0
titleLists = [] #영화제목
countLists = [] #예매율 Lists

for i in range(2010,2020):
    url = "https://www.rottentomatoes.com/top/bestofrt/?year="+i
    res = urlopen(url)
    soup = BeautifulSoup(res, "html.parser")
    li_lists += soup.select("#top_movies_main > div > table > tr > td > a") 


for li in li_lists:
    count += 1
    print(count, li.string.strip())
    countLists.append(count)
    titleLists.append(li.string.strip())
    
# countLists 만으로 DataFrame 생성
df = pd.DataFrame(data=countLists, columns=['Index'])
df['Titles'] = pd.Series(titleLists)
df

# cvs로 저장
df.to_csv("./Data/MoviesList1.csv", sep=",", encoding='utf-8', index=False)
```
